```python
import pandas as pd
```

```python
file_path = '/Users/nancybaraka/Downloads/amz_uk_price_prediction_dataset.csv'
amz_data = pd.read_csv(file_path)
```

```python
product_category_counts = amz_data['category'].value_counts()

top_5_categories = product_category_counts.head(5)

top_5_categories
```

```python
import matplotlib.pyplot as plt


product_category_counts.plot(kind='bar')
plt.show()
```

```python

top_5_categories.plot(kind='pie')
plt.show()
```

```python
mean_price = amz_data['price'].mean()
median_price = amz_data['price'].median()
mode_price = amz_data['price'].mode()[0]  

mean_price, median_price, mode_price
```

```python

variance_price = amz_data['price'].var()
std_dev_price = amz_data['price'].std()

price_range = amz_data['price'].max() - amz_data['price'].min()

Q1 = amz_data['price'].quantile(0.25)
Q3 = amz_data['price'].quantile(0.75)
iqr_price = Q3 - Q1

variance_price, std_dev_price, price_range, iqr_price
```
